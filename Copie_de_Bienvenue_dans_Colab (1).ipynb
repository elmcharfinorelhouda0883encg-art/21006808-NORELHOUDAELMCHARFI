```
import numpy as np
import pandas as pd
import os
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Ignore warnings
warnings.filterwarnings("ignore")

# Define the dataset input path
dataset_input_path = '/kaggle/input/world-education-dataset'

# List files in the dataset's input directory to find the correct filename
print(f"Listing files in {dataset_input_path}:")
try:
    available_files = os.listdir(dataset_input_path)
    for f in available_files:
        print(f)

    # Assuming the main data file is evident from the listing
    # This 'file_to_load' will be updated after seeing the list.
    # For now, we'll try the name that caused the error to demonstrate.
    file_to_load = "world-education-data.csv"

    # Check if a single CSV file is available and use it
    csv_files = [f for f in available_files if f.endswith('.csv')]
    if len(csv_files) == 1:
        file_to_load = csv_files[0]
        print(f"\nDetected single CSV file: {file_to_load}. Using this for loading.")
    elif len(csv_files) > 1:
        print(f"\nMultiple CSV files found: {csv_files}. Please specify which one to load manually.")
        # If multiple CSVs, default to the one expected or the first one found.
        if "world-education-data.csv" in csv_files:
            file_to_load = "world-education-data.csv"
        elif csv_files:
            file_to_load = csv_files[0]

    # Load dataset with the identified file_to_load
    wed_data = pd.read_csv(os.path.join(dataset_input_path, file_to_load))
    print("Dataset loaded successfully. First 5 records:")
    print(wed_data.head())

    # Rest of your code (only execute if dataset loads successfully)
    np.random.seed(0)

    # Fill missing values with mean
    for col in wed_data.select_dtypes(include=[np.number]).columns:
        mean_val = wed_data[col].mean()
        wed_data[col] = wed_data[col].fillna(mean_val)

    # Check for missing values
    print(wed_data.isnull().sum())

    # Label encode the 'country' column
    le = LabelEncoder()
    wed_data['country'] = le.fit_transform(wed_data['country_code'])
    wed_data = wed_data.drop('country_code', axis=1)

    # Visualization: Line chart of primary school enrollment rate over the years
    plt.figure(figsize=(14,6))
    plt.title("World Education Data (1999-2023)")
    sns.lineplot(data=wed_data, x='year', y='school_enrol_primary_pct')
    plt.show()

    # Scatter plot of government expenditure vs. primary school enrollment rate
    sns.scatterplot(x=wed_data['gov_exp_pct_gdp'], y=wed_data['school_enrol_primary_pct'])
    plt.show()

    # Regression plot of government expenditure vs. primary school enrollment rate
    sns.regplot(x=wed_data['gov_exp_pct_gdp'], y='school_enrol_primary_pct', data=wed_data)
    plt.show()

    # Scatter plot with hue for pupil-teacher ratio
    sns.scatterplot(x=wed_data['gov_exp_pct_gdp'], y=wed_data['school_enrol_primary_pct'], hue=wed_data['pupil_teacher_primary'])
    plt.show()

    # Histogram of government expenditure
    sns.histplot(wed_data['gov_exp_pct_gdp'])
    plt.show()

    # Kernel Density Estimation (KDE) plot of government expenditure
    sns.kdeplot(data=wed_data['gov_exp_pct_gdp'], shade=True)
    plt.show()

    # Line plot with dark style
    sns.set_style("dark")
    plt.figure(figsize=(12,6))
    sns.lineplot(data=wed_data, x='year', y='school_enrol_primary_pct')
    plt.show()

    # Correlation heatmap
    sns.heatmap(wed_data.corr(numeric_only=True), annot=True)
    plt.show()

    # Feature Engineering: Calculate the ratio of primary to secondary school enrollment
    wed_data["school_enrol_pri_sec_ratio"] = wed_data["school_enrol_primary_pct"] / wed_data["school_enrol_secondary_pct"]
    print(wed_data.head())

    # Train-test split
    X = wed_data.copy()
    y = X.pop('school_enrol_secondary_pct')
    train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)

    # Random Forest regression model
    forest_model = RandomForestRegressor(random_state=1)
    forest_model.fit(train_X, train_y)

    # Predict and evaluate the model
    y_preds = forest_model.predict(val_X)
    mae = mean_absolute_error(val_y, y_preds)
    print(f"Mean Absolute Error: {mae}")

except FileNotFoundError:
    print(f"Error: Directory not found at {dataset_input_path}. Please ensure the dataset is correctly mounted or its path is accurate.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
